```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Generate sample age data
np.random.seed(42)
ages = np.random.normal(loc=35, scale=8, size=500)

# Create DataFrame
df = pd.DataFrame({'Age': ages})
```

```python
# Initialize scaler
scaler = StandardScaler()

# Fit and transform
df['Age_standardized'] = scaler.fit_transform(df[['Age']])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot original Age
axes[0].hist(df['Age'], bins=30, color='skyblue', edgecolor='black')
axes[0].axvline(df['Age'].mean(), color='red', linestyle='dashed', label='Mean')
axes[0].axvline(df['Age'].mean() + df['Age'].std(), color='green', linestyle='dashed', label='+1σ')
axes[0].axvline(df['Age'].mean() - df['Age'].std(), color='green', linestyle='dashed', label='-1σ')
axes[0].set_title('Original Age Distribution')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')
axes[0].legend()

# Plot standardized Age
axes[1].hist(df['Age_standardized'], bins=30, color='salmon', edgecolor='black')
axes[1].axvline(df['Age_standardized'].mean(), color='red', linestyle='dashed', label='Mean (≈0)')
axes[1].axvline(df['Age_standardized'].mean() + df['Age_standardized'].std(), color='green', linestyle='dashed', label='+1σ')
axes[1].axvline(df['Age_standardized'].mean() - df['Age_standardized'].std(), color='green', linestyle='dashed', label='-1σ')
axes[1].set_title('Standardized Age Distribution')
axes[1].set_xlabel('Z-Score')
axes[1].set_ylabel('Frequency')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Sample data: Age and Salary
np.random.seed(42)
n_samples = 50

ages = np.random.normal(loc=35, scale=8, size=n_samples)
salaries = np.random.normal(loc=65000, scale=12000, size=n_samples)

# Combine into a single dataset
X = np.column_stack([ages, salaries])

# Apply standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Original Data Plot
axes[0].scatter(X[:, 0], X[:, 1], color='blue')
axes[0].set_title('Original Scale: Age vs Salary')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Salary')
axes[0].grid(True)
axes[0].axhline(0, color='black', linewidth=0.5, linestyle='--')
axes[0].axvline(0, color='black', linewidth=0.5, linestyle='--')

# Scaled Data Plot
axes[1].scatter(X_scaled[:, 0], X_scaled[:, 1], color='green')
axes[1].set_title('After Standardization (Z-Scores)')
axes[1].set_xlabel('Standardized Age')
axes[1].set_ylabel('Standardized Salary')
axes[1].grid(True)
axes[1].axhline(0, color='black', linewidth=0.5, linestyle='--')
axes[1].axvline(0, color='black', linewidth=0.5, linestyle='--')

# Equal aspect ratio so we can visually compare scales
axes[0].set_xlim(np.min(ages) - 5, np.max(ages) + 5)
axes[0].set_ylim(np.min(salaries) - 5000, np.max(salaries) + 5000)
axes[1].set_xlim(-3, 3)
axes[1].set_ylim(-3, 3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Generate sample data
np.random.seed(42)
n_samples = 50

ages = np.random.normal(loc=35, scale=8, size=n_samples)
salaries = np.random.normal(loc=65000, scale=12000, size=n_samples)

X = np.column_stack([ages, salaries])

# Apply standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define original axis limits
age_min, age_max = ages.min() - 5, ages.max() + 5
salary_min, salary_max = salaries.min() - 5000, salaries.max() + 5000

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# --- Plot 1: Original Scale ---
axes[0].scatter(X[:, 0], X[:, 1], color='blue')
axes[0].set_title('1. Original Scale: Age vs Salary')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Salary')
axes[0].grid(True)
axes[0].axhline(0, color='black', linewidth=0.5, linestyle='--')
axes[0].axvline(0, color='black', linewidth=0.5, linestyle='--')

# --- Plot 2: After Standardization ---
axes[1].scatter(X_scaled[:, 0], X_scaled[:, 1], color='green')
axes[1].set_title('2. After Standardization (Z-Scores)')
axes[1].set_xlabel('Standardized Age')
axes[1].set_ylabel('Standardized Salary')
axes[1].grid(True)
axes[1].axhline(0, color='black', linewidth=0.5, linestyle='--')
axes[1].axvline(0, color='black', linewidth=0.5, linestyle='--')
axes[1].set_xlim(-3, 3)
axes[1].set_ylim(-3, 3)

# --- Plot 3: Standardized Data on Original Axis Range ---
axes[2].scatter(X_scaled[:, 0], X_scaled[:, 1], color='purple')
axes[2].set_title('3. Standardized Values\non Original Axis Range (Cluttered)')
axes[2].set_xlabel('Standardized Age (on original scale)')
axes[2].set_ylabel('Standardized Salary (on original scale)')
axes[2].grid(True)
axes[2].set_xlim(age_min, age_max)
axes[2].set_ylim(salary_min, salary_max)
axes[2].axhline(0, color='black', linewidth=0.5, linestyle='--')
axes[2].axvline(0, color='black', linewidth=0.5, linestyle='--')

plt.tight_layout()
plt.show()
```
