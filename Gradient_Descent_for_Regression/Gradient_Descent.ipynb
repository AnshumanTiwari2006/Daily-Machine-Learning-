```python
from sklearn.datasets import make_regression
import numpy as np
from sklearn.model_selection import cross_val_score
```

```python
X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=50, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.intercept_
```

```python
np.mean(cross_val_score(reg,X,y,scoring='r2',cv=10))
```

```python
class GradientRegressor:
    def __init__(self, learning_rate, epochs):
        self.m = 100
        self.b = -120
        self.lr = learning_rate
        self.epochs = epochs
    def fit(self,X,y):
        # Calculating the b using GD
        for i in range(self.epochs):
            loss_slope_wrt_b = -2* np.sum(y - self.m*X.ravel() - self.b)
            loss_slope_wrt_m = -2* np.sum((y - self.m*X.ravel() - self.b)*X.ravel())
            self.b = self.b - (self.lr * loss_slope_wrt_b)
            self.m = self.m - (self.lr * loss_slope_wrt_m)
        print(self.m, self.b)

    def predict(self,X):
        return self.m*X + self.b
```

```python
gd = GradientRegressor(0.001, 50) #Even on 100 it works
```

```python
gd.fit(X,y)
```

```python
gd.predict(X)
```
